1. ACTA DE CONSTITUCIÓN

El acta debe tener los datos importantes que son el nombre del proyecto, cliente, lider del equipo de desarrollo, tambien se podría considerar algún encargado legal para ambas parte, el acta de constitución es un documento que formaliza el inicio del proyecto.

Es importante considerar el desplegar el MVP funcional en servidores provistos por la Alcaldía antes de finalizar los 6 meses de plazo, poder lograr la migración exitosa del 100% de los registros iniciales de comerciantes a la nueva base de datos, poder pasar las pruebas de aceptación con un índice de fallos menor en entornos de preproducción.

Algo que se debe considera es la petición de cliente, entre los más importante esta:
* Acceso web multiplataforma.
* Base de datos con respaldo.
* Que pueda ser auditable
* Un sistema de semaforización visual para el estado de licencias sanitarias.

El presupuesto de 70,000 Bs. esta destinado a infraestructura en la nube, licencias de software, viáticos de levantamiento de datos en campo y pago al equipo, por lo que el equipo esta conformado por un lider, un encargado de base de datos, 4 encargados en el desarrollo.

Es importante adjuntar y confirmar los siguientes puntos:
* Aprobación del documento de requerimientos
* Arquitectura base y módulo de usuarios.
* Módulo de cobros funcional.
* Módulo de informes.
* Pruebas en el sistemas
* Despliegue, capacitación y Cierre del Proyecto.

para esto es necesario hacer firmar a todos los encargados o lideres del proyecto para que todos esten conformes con lo que se irá desarrollando en los 6 meses.

2. Documento de Metodología 

La metodología será Ágil, para ello se necesita:
* Product Owner 
* Scrum Master 
* Equipo de Desarrollo 

El proyecto se dividirá en 8 Sprints de 3 semanas cada uno, al finalizar cada sprint es importante una retroalimentación del cliente, para lo que requiere un reunión para la evaluación de los evanaces.

Al escoger la metodología ágil se debe considerar la planeación en la que se define trabajar 8 horas diarias por 5 dias a la semana, luego se debe tomar encuenta las reuniones diarias que debe ser todos los días ya sea manera virtual o presencial esto a coordinación del equipo con una duración entre 15 a 30 minutos dependiendo la lista de tareas, luego se tienen la revisión que será en los cierres de sprint que se mencionó anteriormente, luedo de la revisión se tendra la retrospectiva de la revisión solo los del equipo para ver como mejorar algunos aspectos del proyecto.

Algo que también es importnate es las historias de usuario y considerar cuando se cumplieron en su totalidad, esto se define en base si el código no presenta errores, si el código se encuentra en el github, su también se cumplieron con las pruebas y las validaciones.

3. Solicitud de Proyecto

El comerciantes estan organizados por una mesa directiva, por la cual el presidente y la mesa directiva de los mercados municipales de Sucre con el apoyo de la Alcaldia, realizan los procesos manualmente, esto genera ineficiencias en el registro de puestos y comerciantes, cobros de tarifas, control de licencias sanitarias e informes mensuales para la Dirección de Ingresos retrasos.

Se tiene la idea de poder gestionar mediante un sistema web centralizado que ayude en la recaudación, en la mejor administración de las condiciones de control sanitaria y la dirección de ingresos de datos en tiempo real para la toma de decisiones.

Para continuar se tiene que tener en claro el objetivo que sería desarrollar un sistema web de gestión para Mercados Municipales que digitalice los procesos administrativos del Mercado Campesino, Central y Entre Ríos en un plazo máximo de 6 meses

Para el MVP se considera los siguientes requerimientos:
- Módulo de Usuarios
- Módulo de Puestos
- Módulo de Cobros
- Módulo de Licencias Sanitarias
- Módulo de Reportes

Para el desarrollo también se considera que se tiene como presupuesto 70000 Bs., el plazo de 6 meses con la implementación incluida, la información será proporcionada por la Alcardía.

Se debe considerar importante los puntos que se fueron mencionando, ya que esto permitirá tener una idea más clara de lo que se hará y servirá como un respacto tanto para el equipo de programación como para los encargados de la alcaldia y los comerciantes.

5. Documento de Inception

El principal punto es eliminar el uso de papel, mitigar el riesgo al momento del cobro y todo este respaldado y proveer una herramienta tecnológica adecuadas al proyecto.

Es importante aclara lo que no se hará, no se tendrá compras en línea, no se consideró un sistema de facturación electrónica, como tampoco el control de inventario.

Los interesados en el proyecto son la Alcaldía, los comerciales y la misma población esto tiene que ver más con el cuidado sanitario de cada puesto.

Los problemas más comunes que se puede tener es la falta de conectividad a internet estable, resistencia al cambio tecnológico por parte de algunas personas, pueden ser también algunos de los comerciantes, falta de apoyo en los servidores para el despliegue del proyecto.
